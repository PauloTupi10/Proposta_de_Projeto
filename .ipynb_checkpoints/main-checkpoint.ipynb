```python
# Diferenciação das séries (log-diff)
df_cointegracao['PVPA_diff'] = df_cointegracao['PVPA'].diff().dropna()
df_cointegracao['Media_Mensal_diff'] = df_cointegracao['Media_Mensal'].diff().dropna()
df_cointegracao.dropna(inplace=True)

from statsmodels.tsa.stattools import coint

result_cointegration = coint(df_cointegracao['PVPA_diff'], df_cointegracao['Media_Mensal_diff'], maxlag=1)
# Extraindo os resultados
statistic = result_cointegration[0]
p_value = result_cointegration[1]

if p_value <= 0.05:  # Verifique o valor p
    print(f"As séries são cointegradas com p-value de {p_value}.")
else:
    print("As séries não são cointegradas com p-value de {p_value}.")
```

```python
import statsmodels.api as sm

y = df_cointegracao_interp["PVPA"]
x = df_cointegracao_interp[['Media_Mensal']]

# with statsmodels
x = sm.add_constant(x) # adding a constant

# modelo 
model = sm.OLS(y, x).fit()
parametros = model.params

intercept = parametros["const"]
coef_x = parametros['Media_Mensal'] 


```

```python
# Fazendo uma reta com os parâmetros encontrados
x_RANGE = np.linspace(1.5,17,100)
y = x_RANGE*coef_x+ intercept
y_ultimo = coef_x*df_cointegracao_interp['ARIMA'].values[-1]+ intercept


plt.scatter(df_cointegracao_interp['Media_Mensal'],df_cointegracao_interp['PVPA'])
plt.scatter(df_cointegracao_interp['Media_Mensal'][-1],df_cointegracao_interp['PVPA'][-1],color='red', label="Último Ponto")
plt.scatter(df_cointegracao.loc[datetime(2023,2,28),"Media_Mensal"],
            df_cointegracao.loc[datetime(2023,2,28),"PVPA"],color='green',label="Fevereiro/2023")
# Adicione uma linha vertical em x = 3
plt.axvline(x=df_cointegracao_interp['ARIMA'].values[-1], color='green', linestyle='--', label='Linha Vertical')
plt.axhline(y=y_ultimo, color='red', linestyle='--', label='Linha Horizontal')
plt.plot(x_RANGE,y,color='black',label='Regressão Linear')


plt.xlabel("Expectativa de SELIC")
plt.ylabel("P/VPA")
plt.title("P/VPA x Expectativa de SELIC")
plt.show()

```

```python
def ARIMA_Dados_historicos(serie_valores, intervalo_teste, n_steps = 4):
    from statsmodels.tsa.arima.model import ARIMA
    # intervalo_teste = -12  # -n_steps até 160
    train_data = serie_valores[:-n_steps-intervalo_teste]
    test_data = serie_valores[n_steps+intervalo_teste:-intervalo_teste]
    if (-n_steps-intervalo_teste)==0:
        train_data = serie_valores[:]
        total_data = serie_valores[:]
    elif intervalo_teste<=0:
        total_data = serie_valores[:]
        # Suponha que 'train_data' seja a sua série temporal

        # Primeiro, crie uma cópia do último valor (de 2023-08-31)
        ultimo_valor = total_data.iloc[-1]

        # Crie -intervalo_teste novas datas com frequência mensal a partir do último valor
        novas_datas = pd.date_range(start=total_data.index[-1], periods=-intervalo_teste+1, freq='Q')[1:]

        # Crie uma nova série com os valores do último valor repetidos nas novas datas
        novos_dados = pd.Series([ultimo_valor] * len(novas_datas), index=novas_datas)

        # Concatene a nova série ao final da série original
        total_data = pd.concat([total_data, novos_dados])

    # Agora 'total_data' contém x meses adicionais com o mesmo valor do último

    # Agora 'train_data' contém x meses adicionais com os valores iguais ao de 2023-08-31

    else:
        total_data = serie_valores[:-intervalo_teste]



    (p, d, q) = (4,1,1)
    model = ARIMA(train_data, order=(p, d, q))
    model_fit = model.fit()

    
    # Faça previsões para os próximos 12 meses
    forecast = model_fit.forecast(steps=n_steps)
    
    # Crie um índice de datas para as previsões
    index_forecast = pd.date_range(start=train_data.index[-1], periods=n_steps + 1, freq='Q')[1:]
    data_previsao = train_data.index[-1]
    
    prev_12meses = forecast.sum()

    return data_previsao, prev_12meses
```

```python
# Classifique o DataFrame pelo índice em ordem crescente
df_porAcao_eztec = df_porAcao_eztec.sort_index()

datas_regressao= df_porAcao_eztec.index
df_porAcao_regressao = df_porAcao_eztec["PL"].to_frame().copy()
df_porAcao_regressao["LL"] = 0
#pd.DataFrame(index=datas_regressao, columns=["LL"])
quantidade_dados_regressao = len(datas_regressao)
df_porAcao_regressao.tail(5)
```

```python
import warnings
# Suprimir temporariamente os avisos
warnings.filterwarnings("ignore")
n_steps_number = 4
for intervalo in range(-n_steps_number,quantidade_dados_regressao-6):
    data_previsao, prev_12meses = ARIMA_Dados_historicos(df_porAcao_eztec['LL'], intervalo, n_steps_number)
    df_porAcao_regressao.loc[data_previsao,"LL"] = prev_12meses
```

O patrimônio líquido (PL) da empresa daqui a 12 meses será igual ao PL atual, acrescido do lucro dos próximos 12 meses, permitindo-nos estimar o Valor Patrimonial por Ação (VPA) para o próximo ano.

No entanto, é importante destacar que, em uma análise rigorosa, devemos levar em consideração o desconto dos dividendos. Nesse sentido, o PL futuro é calculado como sendo igual ao PL atual mais o Lucro Líquido (LL) acumulado no período, menos os dividendos distribuídos. É relevante observar que os dividendos representam uma forma distinta de retorno para os investidores e, para a finalidade de mensurar o retorno nos próximos 12 meses, não é necessário descontá-los. O objetivo principal é avaliar o retorno do investimento ao considerar o acréscimo do lucro aos ativos da empresa.

Portanto, podemos estimar o VPA para os próximos 12 meses com base na projeção do PL futuro, considerando o lucro esperado, sem a necessidade de descontar os dividendos, uma vez que estes representam uma maneira diferente de recompensa aos investidores.

No entanto, é importante observar que, para evitar distorções no valor do VPA, especialmente após a data ex-dividendo, é necessário ajustar o VPA diário do ativo por ação. Isso significa que, se a empresa mantiver o mesmo VPA após a data ex-dividendo, o "preço justo" será diferente. Embora esse tipo de empresa geralmente não apresente um Dividend Yield (DY) muito alto, capaz de causar grandes distorções, é importante considerar esses ajustes quando necessário.

Em resumo, a relação entre VPA, expectativas da SELIC e preço pode ser útil para explicar o desempenho de um ativo nos próximos 12 meses. No entanto, é importante ter em mente a necessidade de ajustes no VPA diário em datas ex-dividendo, a fim de evitar distorções no cálculo do "preço justo".
